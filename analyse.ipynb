# Data Analyse

Brug denne notebook til at indlæse, rense og analysere datasæt – ligesom R Markdown.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.formula.api import ols

print('Klar til analyse!')
```

## 1. Indlæs data

Skift filstien til dit eget datasæt (CSV eller Excel).

```python
# CSV:
# df = pd.read_csv('data/mit_datasæt.csv')

# Excel:
# df = pd.read_excel('data/mit_datasæt.xlsx')

# Eksempel-data (slet når du bruger dit eget):
np.random.seed(42)
df = pd.DataFrame({
    'alder': np.random.randint(18, 65, 100),
    'score': np.random.normal(50, 10, 100).round(1),
    'gruppe': np.random.choice(['A', 'B'], 100),
    'indkomst': np.random.normal(35000, 8000, 100).round(0)
})

df.head(10)
```

## 2. Datarensning

```python
# Overblik
print('Størrelse:', df.shape)
print('\nKolonner og typer:')
print(df.dtypes)
print('\nManglende værdier:')
print(df.isnull().sum())
print('\nBeskrivende statistik:')
df.describe().round(2)
```

```python
# Fjern duplikater
df = df.drop_duplicates()

# Fjern rækker med manglende værdier
df = df.dropna()

# Eller erstat manglende med gennemsnit:
# df['score'] = df['score'].fillna(df['score'].mean())

print('Renset datasæt:', df.shape)
```

## 3. Regression (som lm() i R)

```python
# Lineær regression: score ~ alder + indkomst
model = ols('score ~ alder + indkomst', data=df).fit()
print(model.summary())
```

## 4. Plots

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Histogram
axes[0].hist(df['score'], bins=20, color='steelblue', edgecolor='white')
axes[0].set_title('Fordeling af score')
axes[0].set_xlabel('Score')

# Boxplot per gruppe
df.boxplot(column='score', by='gruppe', ax=axes[1])
axes[1].set_title('Score per gruppe')

plt.tight_layout()
plt.show()
```
